```python
'''TAREFA 1
Escreva um programa que solicita que o
usuário digite 5 números inteiros e os coloca
em uma lista.
Quando terminar, imprima a lista.'''

lista_n = []

for i in range(5):
  num1 = int(input(f"Digite o {i+1}° número: "))
  lista_n.append(num1)

print(f"a lista completa é: {lista_n}")
```

```python
'''TAREFA 2
Escreva um programa que aplica a função
módulo a todos os elementos e uma lista.'''

lista2 = [1, -8, 5, 6, -2, 0, 3 ,-7]
lista_absoluta = []

for item in lista2:
  valor_modulado = abs(item)
  lista_absoluta.append(valor_modulado)

print(f"Lista Original: {lista2}")
print(f"Lista com Módulo: {lista_absoluta}")
```

```python
'''TAREFA 5
Faça um programa que solicita uma
quantidade indefinida de números positivos
(deve parar quando um número negativo for
digitado).
Armazene-os em duas listas:
uma para os números pares;
outra para os números ímpares.
A seguir, mostre a porcentagem de pares e
ímpares digitados.'''

pares = []
impares = []

while True:
    numero = int(input("Digite um número positivo (ou um número negativo para parar): "))

    if numero < 0:
        break

    if numero % 2 == 0:
        pares.append(numero)
    else:
        impares.append(numero)

total = len(pares) + len(impares)

if total == 0:
    print("Nenhum número foi digitado.")
else:
    porcentagem_pares = (len(pares) / total) * 100
    porcentagem_impares = (len(impares) / total) * 100

    print("\n===== RESULTADOS =====")
    print(f"Números pares: {pares}")
    print(f"Números ímpares: {impares}")
    print(f"Porcentagem de pares: {porcentagem_pares:.2f}%")
    print(f"Porcentagem de ímpares: {porcentagem_impares:.2f}%")
```

```python
'''TAREFA 6
Faça um programa que solicita uma
quantidade indefinida de números positivos
(deve parar quando um número negativo for
digitado).
Armazene todos os números digitados em
uma lista, sem repetição.'''

numeros = []

while True:
    num = int(input("Digite um número positivo (ou negativo para sair): "))

    if num < 0:
        print("Número negativo digitado. Encerrando...")
        break

    if num not in numeros:
        numeros.append(num)

print("\nNúmeros digitados (sem repetição):", numeros)
```

```python
'''TAREFA 7
Escreva um programa que solicita que o
usuário digite a altura e o peso de 5 pessoas.
Verifique se pelo menos duas das pessoas
tem a mesma altura e mesmo peso.'''

pessoas = []

for i in range(5):
    print(f"\nPessoa {i + 1}:")
    altura = float(input("Digite a altura (em metros): "))
    peso = float(input("Digite o peso (em kg): "))
    pessoa = (altura, peso)
    pessoas.append(pessoa)

if len(pessoas) != len(set(pessoas)):
    print("\nPelo menos duas pessoas têm a mesma altura e o mesmo peso.")
else:
    print("\nNenhuma pessoa tem altura e peso iguais.")
```

```python
'''TAREFA 8
Faça um programa que solicita que o
usuário digite o nome e a idade de diversas
pessoas (o programa deve parar quando
um nome vazio for informado).
Retorne o nome da pessoa mais velha.'''

pessoas = []

while True:
    nome = input("Digite o nome (ou pressione Enter para sair): ")
    if nome == "":
        break
    idade = int(input("Digite a idade: "))
    pessoas.append((nome, idade))

if pessoas:
    mais_velha = max(pessoas, key=lambda x: x[1])
    print(f"\nA pessoa mais velha é {mais_velha[0]} com {mais_velha[1]} anos.")
else:
    print("\nNenhuma pessoa foi cadastrada.")
```

```python
'''TAREFA 2
Dado o dicionário ao lado:

Imprima o total de episódios
Substitua a quantidade de episódios pelo
percentual (ep_anime/ep_total)
Imprima nome e percentual de cada
anime, em ordem decrescente de
percentual

animes = { "Naruto": 220,
"Jujutsu Kaisen": 47, "Dragon Ball Z": 291, "Death Note": 37,
"Fullmetal Alchemist": 64,
"Evangelion": 26,
"Berserk": 25,
"Code Geass": 50, "Akame ga Kill!": 24, "Elfen Lied": 13
}'''

animes = {
    "Naruto": 220,
    "Jujutsu Kaisen": 47,
    "Dragon Ball Z": 291,
    "Death Note": 37,
    "Fullmetal Alchemist": 64,
    "Evangelion": 26,
    "Berserk": 25,
    "Code Geass": 50,
    "Akame ga Kill!": 24,
    "Elfen Lied": 13
}

ep_total = sum(animes.values())
print("Total de episódios:", ep_total)

for anime in animes:
    animes[anime] = animes[anime] / ep_total * 100

animes_ordenados = sorted(animes.items(), key=lambda x: x[1], reverse=True)

print("\nPercentual de episódios por anime (ordem decrescente):")
for nome, percentual in animes_ordenados:
    print(f"{nome}: {percentual:.2f}%")
```
