```python
alunos_dp = [
    {"matricula": "123", "nome": "Lucas", "curso": "ADS"},
    {"matricula": "124", "nome": "Luan", "curso": "Administração"},
    {"matricula": "125", "nome": "Leandro", "curso": "Direito"},
    ]

print("1. Sistema Simplificado de Cadastro de Alunos")
print("-" * 40) # Fixed the concatenation error

#1.Adicionar novo aluno...(CREATE)
print("-> Adicioando novo aluno...")
novo_aluno = {"matricula": "126", "nome": "Fernanda", "curso": "Engenharia Civil"} # Fixed the syntax error

#2.Buscar Aluno por Mtricula (READ)
matricula_busca = "124"
aluno_encontrado = next((aluno for aluno in alunos_dp if aluno["matricula"] == matricula_busca), None)



if aluno_encontrado:
  print(f"\nAluno encontrado (Matricula: {matricula_busca}):")
  # Acessa os valores do dicionario usando chaves.
  print(f"Nome: {aluno_encontrado}")
else:
  print(f"\nAluno com matricula {matricula_busca} não encontrado.")

# Atualizar informação do curso (Update)
matricula_update = "2023003"
for aluno in alunos_dp:
  if aluno["matricula"] == matricula_update:
    aluno["curso"] = "Segurança da Informação"
    print(f("\nCurso do aluno (aluno['nome'])) atualizado para {aluno['curso']}.")
    break
print("-"*40)
```
