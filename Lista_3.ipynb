```python
#15. Elabore um algoritmo que solicite N números reais e quando o usuário informar o valor
#nulo 0 (zero) o programa ordene e mostre todos os números informados de forma crescente

n15 = []
n15.append(int(input("Digite um numero: ")))
ultimo_elemento=n15[-1]

while ultimo_elemento!=0:
  n15.append(int(input("Digite um numero: ")))
  ultimo_elemento=n15[-1]

n15.sort()
print(f"os numeros em ondem cresente são: {n15}")
```

```python
#16. Escreva um programa que vá solicitando as idades dos alunos da sala até que todos
#sejam informados (perguntar ao usuário se deseja informar a idade do próximo aluno). Ao
#final apresentar a idade do mais novo, a idade do mais velho, Quantos alunos têm mais de
#18 anos, quantos alunos têm até 18 anos, a média aritmética e a mediana.

idades = []
continuar = 'S'
contador = 1
soma_idades = 0
maiores_de_18 = 0

print("--- Coleta de Idades dos Alunos ---")

while continuar == 'S':
    idade_input = input(f"Informe a idade do Aluno #{contador}: ")
    idade = int(idade_input)

    if idade <= 0:
        print("Idade inválida. Vamos tentar o próximo.")
    else:
        idades.append(idade)
        soma_idades += idade
        contador += 1

        if idade > 18:
            maiores_de_18 += 1

    continuar = input("Deseja informar a idade do próximo aluno? (S/N): ").upper().strip()
    print("-" * 20)

total_alunos = len(idades)

if total_alunos == 0:
    print("\nNenhuma idade informada.")
else:
    idade_mais_novo = min(idades)
    idade_mais_velho = max(idades)

    ate_18 = total_alunos - maiores_de_18

    media_aritmetica = soma_idades / total_alunos

    idades.sort()

    meio = total_alunos // 2

    if total_alunos % 2 == 1:
        mediana = idades[meio]
    else:
        elemento1 = idades[meio - 1]
        elemento2 = idades[meio]
        mediana = (elemento1 + elemento2) / 2

    print("\n\n" + "="*35)
    print("RELATÓRIO FINAL DE ESTATÍSTICA")
    print(f"Total de alunos processados: {total_alunos}")
    print("="*35)
    print(f"Idade do MAIS NOVO: {idade_mais_novo} anos")
    print(f"Idade do MAIS VELHO: {idade_mais_velho} anos")
    print("-" * 35)
    print(f"Alunos com MAIS de 18 anos: {maiores_de_18}")
    print(f"Alunos com ATÉ 18 anos: {ate_18}")
    print("-" * 35)
    print(f"MÉDIA das idades: {media_aritmetica:.2f} anos")
    print(f"MEDIANA das idades: {mediana:.2f} anos")
    print("="*35)

```
