```python
#16. Escreva um programa que solicite o valor fixo do salário de um vendedor, o total vendido no mês e o percentual de comissão do vendedor. Ao final apresentar o salário bruto.
salario = float(input("Digite o valor do seu salário fixo: "))
totalVendas = float(input("Digite o total de vendas no mês: "))
percentual = float(input("Digite o percentual de comissão: "))

comissão = totalVendas * (percentual / 100)
salarioBruto = salario + comissão

print("O salário bruto é: R$", salarioBruto)
```
